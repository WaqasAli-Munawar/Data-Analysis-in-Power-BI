Microsoft Power BI gives us a wide range of methods to analyze our data and get answers to our questions. This module examines the Power BI features that are based in advanced analytics and visualization.

Occasionally, the fastest way to get an answer is to ask a question by using our own words. Power BI allows us to do exactly that method with its advanced analytics AI capabilities. One feature of Power BI is that it allows us to ask questions by using natural language, and then it will answer those questions for us.

Similarly, our dataset is likely numerical, where it consists of numbers, amounts, measures, and so on. We've been successfully analyzing this numerical data to get insights. Analyzing non-numerical data can be difficult, but with the AI capabilities of Power BI, we can analyze text data to get more insights than before. For example, we might have an abundance of comments or reviews from customers or the results of an employee survey that have been left long forgotten in a database. Now, we can analyze this additional data and transform it into valuable information.

Because the numbers don't tell us everything, we can use the AI features in Power BI to examine our data further, to see what people are saying, and to get more constructive and meaningful results, which will help us ultimately make better business decisions.

For this lesson's scenario, we have been building reports for different teams in our organization. The **Q&A feature** in Power BI lets us explore our data in our own words by allowing us to ask natural language questions and then providing us with answers to those questions.

This ability to ask questions is valuable to us, as the report author, and to our report users. It gives us ideas for the type of visuals that we can display in our report and lets us quickly add those visuals. Additionally, it gives our report users an effective tool that they can use to get quick answers to their questions about the data, independently. This self-help aspect to Power BI saves time for everyone involved.

Power BI records all questions that are asked, and we can use this information to set up the Q&A feature to be more effective. When the Q&A feature answers so many questions, we'll have fewer people turning to us for those answers.

Consider a situation where we've created a report for the Supply Chain team, and now the team members have questions about various views or insights that they are interested in. We have become inundated with questions and don't have time to address each one individually.

Consequently, we have decided to implement the Q&A feature so that users can ask questions and get answers by themselves.

**Add the Q&A visual to our report canvas**

To get access to the Q&A feature, we need to add the Q&A visual to our report. We can double-click anywhere on the canvas, and the visual should appear. Alternatively, we can select the **Q&A icon** on the **Visualizations pane**.



The Q&A feature is also available as a button, which is a useful option if we want to save space on our report canvas.



When the Q&A visual or button is added to our report, we can reposition and resize it. We can also customize the formatting in the same way that we would for any other type of visual or button.

We can start asking questions immediately by selecting one of the suggested questions or by entering a question into the question box. As we type, Power BI will automatically display suggestions to help us complete our question.



This lesson uses real data from Dataquest. The dataset contains records about screen completion rate by lesson. Details about each field (column) are as follows:

* **completed:** the number of learners that completed the screen.
* **language:** the programming language taught in the lesson (SQL, R, Python, etc.).
* **mission:** the "lesson" number. Dataquest now uses the term "lesson" site-wide, instead of "mission".
* **mode:** provides additional information about the type of screen.
* **screen:** the screen number.
* **status:** publishing status.
* **students:** the number of learners that viewed the screen.
* **type:** the lesson type. mission refers to a standard lesson at Dataquest, used for teaching coding in Python, SQL, R, etc.

**Set up the Q&A feature**

When we have added the **Q&A visual** to our report, we can set up the underlying Q&A feature so that it improves in answering questions about our data. Basically, we will teach the Q&A feature to better understand people. This setup can be useful from the outset so that we can get the visual ready for active use. However, setup does not stop at that point; we can proactively monitor and review the questions that are coming through from users, and then we can address misunderstandings or common typos. We can also manage the key terms that are associated with our data, so we can add a library of synonyms that might be entered by different users across the organization when they are asking questions about the data. We can constantly calibrate the Q&A feature so that it provides better answers to our organization's questions.

In this example, we receive feedback from users saying that they can't get data on sales by country. To assess the problem, we enter sales by country into the question box. The visual does not update; it does not answer your question. The following image shows that the word country is underlined in red. When a red underline occurs, Power BI is telling us that it does not understand this term. Knowing that the word country is not used in our dataset, we decide to use the term region instead and ultimately discover why the question is not being answered. Accordingly, we realize that we need to teach Power BI what we mean by adding a new term to its thesaurus.



* Select the **settings icon** to the right of the question box to open the **Q&A setup window**. 
* Then, select the **Teach Q&A** option.



Enter our question again, and then select the **Submit** button. In the **Define the terms Q&A didn't understand** section that displays, enter our alternative term or synonym. In this case, we will enter region. The following image shows that Power BI displays a preview result so that we can see if this new term will return the results that we are looking for. If this result is correct, select **Save**.



Now, when users search for sales by country, Power BI will know that they really mean sales by region and will automatically display the associated data in the visual.



Let's enter a question in our Q&A visual.

**Use the Q&A feature to build visuals**

When Power BI answers a question, and we find the visual result to be engaging or helpful, we can add it as a standard visual to our report. For example, if we review the questions that are being asked, and we see that several users are asking the same question, we can add the answer to this question as a standard visual in the report so they no longer have to ask the question. Similarly, we can also use the Q&A feature to start building our report by asking questions and adopting the suggested visual result formats from Power BI.

To turn a Q&A result into a standard visual, **select the icon** next to the question box.



The Q&A feature is unique in that it does not require users to have knowledge of Power BI to use the visual; users can ask their questions and they, too, can create insightful visuals.

The **Key influencers visual** helps us understand the factors that are affecting a specific metric. It analyzes our data for us, ranks the factors that matter, and then displays those factors as key influencers. The visual also helps us to contrast the relative importance of these factors, meaning that we can build our visuals while understanding what factors impact those visuals and why the visuals appear the way that they do.

In this example, we've built several visuals for the Customer Service team, and now we want to understand the factors that most influence our metrics. Specifically, we want to figure out what factors are sales. One factor might be the client type and another might be location, but we're uncertain. The Key influencers visual will find that information for us.

To establish the key influencers,
* first add the Key influencers visual to our report by selecting the **Key influencers icon** on the **Visualization** pane. 
* Then, populate the visual with the metrics that we want to measure. In this case, we're interested in sales, so in the **Analyze** field well, we will add the `Sales` field, and in the **Explain by** field well, we will add the `City` and `Product Name` fields. 
* The visual updates according to the fields that we added and then shows the influence that those fields have on our data.



We can now use the **What influences**... drop-down list to see what caused the data to decrease or increase. The previous image shows that the sales for California are likely to be USD 130,700.00 higher than sales in other cities. For more information, see [Create key influencers visualizations](https://docs.microsoft.com/en-us/power-bi/visuals/power-bi-visualization-influencers).

The **Decomposition Tree visual** automatically aggregates our data and lets us drill down into our dimensions so that we can view our data across multiple dimensions. Because Decomposition Tree is an AI visual, we can use it for improvised exploration and conducting root cause analysis.

In this example, we've built visuals for the Supply Chain team, but the visuals do not answer all the team's questions. In particular, the team wants to be able to analyze the percentage of products that the organization has on back order, in other words, the percentage of products that are out of stock. The Decomposition Tree visual can help us accomplish that task.

Add the Decomposition Tree visual to our report by selecting the **Decomposition Tree icon** on the **Visualization** pane. Then, in the **Analyze** field well, add the measure or aggregate that we want to analyze. In the **Explain by** field well, add the dimension(s) that we want to drill down into. In this case, we want to analyze the `Sales` field by drilling down into a number of dimensions, such as `Country`, `City`, and `Product`, as illustrated in the following image.



The visual updates according to the fields that we added and displays the analysis summary result. In this case, the value of sales is USD 13,499,680.00. We can select the plus (+) sign, which will present the drill-down options that we have added. We can select any of the fields in the drop-down list to drill down into the data and see how it contributed to the overall result.

At the top of the list of dimensions that we added are two additional options that are marked with **lightbulb icons**. These options are referred to as AI splits, and they'll automatically find high and low values in the data for us.



AI splits work by considering all available fields and determining which one to drill into to get the highest/lowest value of the measure that is being analyzed. We can use the results of these splits to find out where we should look next in the data. The following image illustrates the result of selecting the High value AI split.



For more information, see [Create and view decomposition tree visuals in Power BI](https://docs.microsoft.com/en-us/power-bi/visuals/power-bi-visualization-decomposition-tree).
